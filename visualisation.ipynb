```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("E:\movie imdb\players.csv")
```

```python
#plotting Players and Runs on graph
xpoints=df['Player']
ypoints=df['Runs']

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#plotting the entire dataframe using seaborn liabrary
sns.displot(df)
```

```python
#classifying players on bases on players average
sns.regplot(x='Runs', y='Ave', data=df)
plt.show()
```

```python
#classifying players on bases on players balls faced in entire career
sns.regplot(x='Runs', y='BF', data=df)
plt.show()
```

```python
#classifying players on bases on players strike rate in entire career
sns.regplot(x='Runs', y='SR', data=df)
plt.show()
```

```python
#classifying players on bases on players century[100's] in entire career
sns.regplot(x='Runs', y='100', data=df)
plt.show()
```

```python
#classifying players on bases on players half-century[50's] in entire career
sns.regplot(x='Runs', y='50', data=df)
plt.show()
```

```python
#classifying players on bases on players balls faced in entire career through histogram repersentation
sns.histplot(x='Runs', y='BF', bins=10, data=df)
plt.show()
```

```python
#classifying players on bases on players average in entire career through histogram repersentation
sns.histplot(x='Runs', y='Ave', kde=20, bins=30, hue='SR', data=df)
plt.show()
```
